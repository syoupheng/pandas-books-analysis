### Le format des dates n'est pas uniforme et si l'on souhaite faire des calculs avec les dates il faut convertir les dates en "float".

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
```

```python
df.loc[1905:, 'Date of Publication'].head(10)
```

```python
df['Place of Publication'].head(10)
```

```python
df.loc[4157862]
```

```python
df.loc[4159587]
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
```

```python
try:
    df.to_sql("Books", engine, if_exists="append", index=True, index_label="id", dtype={"Date of Publication":Float()})
except (exc.IntegrityError, exc.OperationalError) as err:
    print(err.orig.args)
```
